```python
import numpy
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# import embeddings
data = numpy.genfromtxt("embed.csv", delimiter=',')
data = data[1:,1:] # remove header and key
```

```python
# Generate the linkage matrix
Z = linkage(data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
agg_clustering = AgglomerativeClustering(n_clusters=40)
y_pred = agg_clustering.fit_predict(data)
y_pred
```

```python
agg_clustering.n_clusters
```

```python
# for viz only
pca = PCA(n_components=2)
data_r = pca.fit(data).transform(data)
```

```python
plt.scatter(data_r[:, 0], data_r[:, 1], c=y_pred, cmap='rainbow')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
